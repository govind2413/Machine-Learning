```python
#Next step of EDA : To handle categorical data
#In this dataset every column contains object type data so we will apply LabelEncoder on this dataset
```

```python
#Applying LabelEncoder
from sklearn.preprocessing import LabelEncoder
#creating object of LabelEncoder class 
le=LabelEncoder()
```

```python
df_new=df
```

```python
#Applying LabelEncoder by using for loop
for i in df_new:
    df_new[i]=le.fit_transform(df[i])
```

```python
df_new
```

```python
#We will now seperate input and output columns from dataset 
x=df_new.drop(['class'],axis=1)
y=df_new['class']
```

```python
#Features/ input columns 
x.columns
```

```python
#output columns
y
```

```python
#Next step of EDA: Feature scaling 
#before that we will split data on 70% training and 30% testing 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train.shape,x_test.shape
```

```python
#Applying StandardScaler on Training input and testing ouput
from sklearn.preprocessing import StandardScaler
#creating object of StandardScaler class
ss=StandardScaler()
```

```python
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

```python
x_train
```

```python
type(x_train),x_train.ndim
```

```python
type(x_test),x_test.ndim
```

```python
#we have binary classifier dataset we will create the model with the help of LogicticRegression
from sklearn.linear_model import LogisticRegression
#creating the object of LogisticRegression class 
lr=LogisticRegression()
```

```python
##now we will train the model with 70% data which is x_train and y_train output
#use inbuilt inbuilt method fit()
lr.fit(x_train,y_train)
```

```python
#we will test the model with 30% data, using inbuilt method predict()
y_pred=lr.predict(x_test)
y_pred
```

```python
#Creating a dictionary
dict={'Actual Output':y_test,'Predicted Output':y_pred}
df2=pd.DataFrame(dict)
df2
```

```python
#we will now elavaluate the model using confusion_matrix algorithm
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred))
```

```python
#we will check the classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#As observed in the above report, the recall score is an important metric. A score of 0.94 indicates a 94% recall rate, which is satisfactory. 
#Similarly, the accuracy is 0.95, representing 95%, which is also considered good. We will implement additional algorithms 
#to evaluate which one provides the best recall score and accuracy.
```
